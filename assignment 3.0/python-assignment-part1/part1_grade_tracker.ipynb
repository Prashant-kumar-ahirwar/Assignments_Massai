# part 2

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
# 1. PRINT SUBJECT + MARKS + GRADE
print("Subject-wise Performance:\n")
a=-1
for i in marks:
    a+=1
    # Grade logic
    if i >= 90:
        grade = "A+"
    elif i >= 80:
        grade = "A"
    elif i >= 70:
        grade = "B"
    elif i >= 60:
        grade = "C"
    else:
        grade = "F"
    print(f"{subjects[a]} : {i} → {grade}")
```

```python
# 2. CALCULATIONS
total = sum(marks)
average = round(total / len(marks), 2)

# Highest & Lowest
max_marks = max(marks)
min_marks = min(marks)

highest_subject = subjects[marks.index(max_marks)]
lowest_subject = subjects[marks.index(min_marks)]

print("\n--- Summary ---")
print("Total Marks   :", total)
print("Average Marks :", average)
print(f"Highest       : {highest_subject} ({max_marks})")
print(f"Lowest        : {lowest_subject} ({min_marks})")
```

```python
# 3. WHILE LOOP (ADD NEW SUBJECTS)
new_subjects = []
new_marks = []
while True:
    sub = input("Enter subject name (or type 'done' to stop): ")

    if sub.lower() == "done":
        break

    mark_input = int(input("Enter marks (0-100): "))

    if mark_input < 0 or mark_input > 100:
        print("Invalid marks! Must be between 0 and 100.")
        continue

    new_subjects.append(sub)
    new_marks.append(mark_input)

# Merge old + new lists
subjects.extend(new_subjects)
marks.extend(new_marks)
print(f"{subjects}\n {marks}")
# Final summary
print("\nNew subjects added:", len(new_subjects))

if len(marks) > 0:
    new_avg = round(sum(marks) / len(marks), 2)
    print("Updated Average:", new_avg)
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
passed_student = 0
failed_students = 0
max_mark = 0
class_total_marks = 0
print(f"{'Name':<15} | {'Average':<7} | {'Status':<5}")
print("-"*35)
for i in class_data:
    a = sum(i[1])/len(i[1])
    if max_mark<a:
        max_mark = a
        topper = i[0]
    if a >= 60:
        r = "pass"
        passed_student = passed_student+1
    else:
        r = 'fail'
        failed_students = failed_students + 1
    print(f"{i[0]:<15} | {a:<7} | {r:<5}")
    
    class_total_marks = class_total_marks + a

print("-"*35, end='\n\n')
print("Passed Students :", passed_student)
print("Failed Students :", failed_students)
print(f"Topper          : {topper}({max_mark})")
print(f"class average   : {round(class_total_marks / len(class_data), 2)}")
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

```python
# 1. Strip remove extra whitespaces on first and last
clean_essay = essay.strip()
print("1. Clean Essay:")
print(clean_essay, end='\n\n')

# 2. Title Case
title_case = clean_essay.title()
print("2. Title Case:")
print(title_case, end='\n\n')

# 3. Count "python"
count_python = clean_essay.count("python")
print("3. Count of 'python':", count_python, end='\n\n')

# 4. Replace "python"
clean_essay = clean_essay.replace("python", "Python 🐍") # changes on our cleaned file
print("4. After Replacement:")                     # if we use diff. variable name no change happens to clean_essay
print(clean_essay, end='\n\n')

# 5. Split into sentences
sentences = clean_essay.split(". ")
print("5. Sentence List:")
print(sentences, end='\n\n')

# 6. Print numbered sentences
for i in range(len(sentences)):
    if sentences[i][-1] != ".":
        print(f"{i+1}-> {sentences[i]}.")
    else:
        print(f"{i+1}-> {sentences[i]}")

```
