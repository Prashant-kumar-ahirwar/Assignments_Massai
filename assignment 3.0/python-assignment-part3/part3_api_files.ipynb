```python
import requests

BASE_URL = "https://dummyjson.com/products"

# Utility: Safe API Request

def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"[ERROR] GET request failed: {e}")
        return None


def safe_post(url, data):
    try:
        response = requests.post(url, json=data, timeout=5)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"[ERROR] POST request failed: {e}")
        return None

```

```python

# Step 1 — Fetch & Display

def fetch_products():
    print("\n--- Fetching Products ---")
    url = f"{BASE_URL}?limit=20"
    data = safe_get(url)

    if not data:
        return []

    products = data.get("products", [])

    print(f"\n{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<10} | {'Rating'}")
    print("-" * 80)

    for p in products:
        print(f"{p['id']:<4} | {p['title'][:30]:<30} | {p['category']:<15} | ${p['price']:<9} | {p['rating']}")

    return products
```

```python
# -------------------------
# Step 2 — Filter & Sort
# -------------------------
def filter_and_sort(products):
    print("\n--- Filtered (Rating ≥ 4.5) & Sorted by Price DESC ---")

    filtered = [p for p in products if p.get("rating", 0) >= 4.5]

    if not filtered:
        print("No products match the criteria.")
        return

    sorted_products = sorted(filtered, key=lambda x: x["price"], reverse=True)

    for p in sorted_products:
        print(f"{p['title']} | ${p['price']} | Rating: {p['rating']}")
```

```python
# -------------------------
# Step 3 — Category Search
# -------------------------
def fetch_laptops():
    print("\n--- Laptop Products ---")

    url = f"{BASE_URL}/category/laptops"
    data = safe_get(url)

    if not data:
        return

    products = data.get("products", [])

    for p in products:
        print(f"{p['title']} | ${p['price']}")
```

```python
# -------------------------
# Step 4 — POST Request
# -------------------------
def create_product():
    print("\n--- Creating Product (Mock API) ---")

    url = f"{BASE_URL}/add"

    payload = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    response = safe_post(url, payload)

    if response:
        print("Response from server:")
        for key, value in response.items():
            print(f"{key}: {value}")
```

```python
# -------------------------
# MAIN EXECUTION
# -------------------------
if __name__ == "__main__":
    products = fetch_products()
    
    if products:
        filter_and_sort(products)
    
    fetch_laptops()
    create_product()
```

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


# Test cases
print(safe_divide(10, 2))      # ✅ 5.0
print(safe_divide(10, 0))      # ❌ divide by zero
print(safe_divide("ten", 2))   # ❌ invalid type
```

```python
def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()  # handles HTTP errors like 404/500
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e}")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None


def safe_post(url, data):
    try:
        response = requests.post(url, json=data, timeout=5)
        response.raise_for_status()
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e}")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None
```

```python
def product_lookup_loop():
    while True:
        user_input = input("\nEnter a product ID to look up (1–100), or 'quit' to exit: ").strip()

        if user_input.lower() == "quit":
            print("Exiting...")
            break

        # Validate integer input
        if not user_input.isdigit():
            print("Invalid input. Please enter a number.")
            continue

        product_id = int(user_input)

        if not (1 <= product_id <= 100):
            print("Invalid range. Enter a number between 1 and 100.")
            continue

        # API Call
        url = f"https://dummyjson.com/products/{product_id}"

        try:
            response = requests.get(url, timeout=5)

            if response.status_code == 404:
                print("Product not found.")
            elif response.status_code == 200:
                data = response.json()
                print(f"Title: {data['title']}")
                print(f"Price: ${data['price']}")
            else:
                print(f"Unexpected status code: {response.status_code}")

        except requests.exceptions.ConnectionError:
            print("Connection failed. Please check your internet.")
        except requests.exceptions.Timeout:
            print("Request timed out. Try again later.")
        except Exception as e:
            print(f"Unexpected error: {e}")


# Run loop
product_lookup_loop()
```

```python
from datetime import datetime

LOG_FILE = "error_log.txt"

# -------------------------
# Logger Function
# -------------------------
def log_error(function_name, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"
    
    try:
        with open(LOG_FILE, "a", encoding="utf-8") as file:
            file.write(log_entry)
    except Exception as e:
        print(f"Logging failed: {e}")


# -------------------------
# Trigger 1 — ConnectionError
# -------------------------
def trigger_connection_error():
    print("\n--- Triggering Connection Error ---")
    
    url = "https://this-host-does-not-exist-xyz.com/api"
    
    try:
        requests.get(url, timeout=5)
    except requests.exceptions.ConnectionError as e:
        print("Connection error occurred.")
        log_error("fetch_products", "ConnectionError", str(e))
    except Exception as e:
        log_error("fetch_products", "UnexpectedError", str(e))


# -------------------------
# Trigger 2 — HTTP 404 Error
# -------------------------
def trigger_http_error():
    print("\n--- Triggering HTTP 404 Error ---")
    
    product_id = 999
    url = f"https://dummyjson.com/products/{product_id}"
    
    try:
        response = requests.get(url, timeout=5)
        
        if response.status_code != 200:
            message = f"{response.status_code} Not Found for product ID {product_id}"
            print("HTTP error detected.")
            log_error("lookup_product", "HTTPError", message)
        else:
            print("Unexpected success (product exists).")
    
    except requests.exceptions.ConnectionError as e:
        log_error("lookup_product", "ConnectionError", str(e))
    except requests.exceptions.Timeout as e:
        log_error("lookup_product", "Timeout", str(e))
    except Exception as e:
        log_error("lookup_product", "UnexpectedError", str(e))


# -------------------------
# Read and Display Logs
# -------------------------
def display_logs():
    print("\n--- Error Log Contents ---")
    
    try:
        with open(LOG_FILE, "r", encoding="utf-8") as file:
            content = file.read()
            print(content if content else "Log file is empty.")
    except FileNotFoundError:
        print("No log file found yet.")


# -------------------------
# MAIN
# -------------------------
if __name__ == "__main__":
    trigger_connection_error()
    trigger_http_error()
    display_logs()
```
